```python
import numpy as np
import matplotlib.pyplot as plt
```

### Global Matplotlib Formatting

```python
# LaTeX for plot labels
plt.rcParams['text.usetex'] = True

# Specify the LaTeX package for standard math fonts
plt.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'


plt.rcParams['xtick.labelsize'] = 15  # Example for x-axis tick labels
plt.rcParams['ytick.labelsize'] = 15 
```

### Magnetic field contours and field strengh

```python

x = np.linspace(-0.5, 0.5, 400)
z = np.linspace(-0.5, 0.5, 400)
X, Z = np.meshgrid(x, z)


# Calculate the flux function values
psi_0 = (np.cos(np.pi * X) * np.sin(2 * np.pi * np.abs(Z))) / (2 * np.pi)

# Calculate the magnetic field strength 
B0_x = -np.cos(np.pi * X) * np.cos(2 * np.pi * Z)  # x-component of B_0
B0_z = -0.5 * np.sin(np.pi * X) * np.sin(2 * np.pi * Z)  # z-component of B_0
B_strength = np.sqrt(B0_x**2 + B0_z**2)

# Plot the contours of psi_0
plt.figure(figsize=(10, 8))
contour = plt.contour(X, Z, psi_0, colors='k')

clabels = plt.clabel(contour, inline=True, fontsize=18)
for label in clabels:
    label.set_fontweight('bold')

# Plot the magnetic field strength in the background
plt.imshow(B_strength, extent=(-0.5, 0.5, -0.5, 0.5), origin='lower', cmap='magma', alpha=0.9)


# Format
cbar = plt.colorbar(label='Magnetic Field Strength')
cbar.set_label(r'Magnetic Field Strength $\left|\boldsymbol{B}_0\right|$', size=20) 
plt.xlabel(r'$x$',size=25)
plt.ylabel(r'$z$',size=25)
plt.title(r'Initial magnetic field lines and field strength $\left|\boldsymbol{B}_0\right|$', size=22, pad=10)

plt.savefig('initial_magnetic_field.png')
plt.show()
```
