## Phase Diagram 

```python
SL = [10**3,10**4,10**5]
SL = 0.5 * [SL]
LI = [0.01, 0.0175, 0.045, 0.1, 1]
LI = [1/l for l in LI]
```

```python

# Create all combinations of array1 and array2
X, Y = np.meshgrid(SL, LI)

# Flatten the arrays to get all combinations as x and y values for plotting
x_vals = X.flatten()
y_vals = Y.flatten()



# Generate y values for plotting x = y^2
y_range = np.linspace(0, 10**3, 400)  # Adjust range and density as needed
x_from_y_squared = y_range**2

# Plotting
plt.figure(figsize=(10, 6))
plt.loglog(x_vals, y_vals, 'x', markersize=10, color='k')  # Plot as 'X's on a log-log scale
plt.loglog(x_from_y_squared, y_range, '-', label='$x=y^2$')
plt.ylabel(r'$\frac{1}{\lambda_i}$', size=25, rotation =360, labelpad=25)
plt.xlabel(r'$S_L$', size=20)
plt.title('Combinations of Two Arrays on a Log-Log Scale')

plt.show()
```
