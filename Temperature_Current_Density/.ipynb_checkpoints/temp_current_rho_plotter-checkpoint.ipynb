```python
import os as os 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import sdf_helper as sh 
import matplotlib.colors as mcolors 
from matplotlib.ticker import MaxNLocator 
import glob as glob
%matplotlib inline 
from mpl_toolkits.axes_grid1 import make_axes_locatable 
```

```python
hard_drive = '/volumes/My Passport/MMath Project/sim_800_15'
os.chdir(hard_drive)


data_dir = hard_drive + '/Data/' 
data = sh.getdata(0,data_dir, verbose=False) 
```

```python
# LaTeX for plot labels
plt.rcParams['text.usetex'] = True

# Specify the LaTeX package for standard math fonts
plt.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'


plt.rcParams['xtick.labelsize'] = 20  # Example for x-axis tick labels
plt.rcParams['ytick.labelsize'] = 20 
```

```python
def plot_stacked_horizontal(ds, data_dir='Data'): 
    n = len(ds) # Number of datasets 

    # Creating a grid of subplots
    fig, axs = plt.subplots(n, 3, figsize=(17, 3 * n), sharex='col', sharey='row') 
    
    # Ensuring axs is always a 2D array for consistency in indexing 
    if n == 1: 
        axs = axs.reshape(1, -1) 

    # Define the quantities to plot 
    current_types = ['Fluid_Temperature', 'Current_Jz', 'Fluid_Rho'] 
    current_labels = [r'Temperature $T$', r'Current $J_y$', r'Density $\rho$'] 
    colorbar_labels = [r'$T$', r'$J_y$', r'$\rho$']
    # Define Colour Maps
    maps = ['hot', 'magma', 'viridis'] 

    vmin_vmax = {ctype: (np.inf, -np.inf) for ctype in current_types} # Initialise vmin, vmax 

    # Pre-calculate global vmin and vmax for each type 
    for d in ds: 
        data = sh.getdata(d, data_dir, verbose=False) 
        for ctype in current_types: 
            current_data = getattr(data, ctype).data 
            vmin, vmax = vmin_vmax[ctype] 
            vmin_vmax[ctype] = (min(vmin, np.min(current_data)), max(vmax, np.max(current_data))) 


    for i, dataset in enumerate(ds): 
        # Load data 
        data = sh.getdata(dataset, data_dir, verbose=False) 
        t = data.Last_dump_time_requested.data 

        # Extract grid for plotting 
        g = data.Grid_Grid 
        xb = g.data[0] 
        yb = g.data[1] 

        X, Y = np.meshgrid(xb, yb, indexing='ij') 
        for j, current_type in enumerate(current_types): 
            # Selecting the correct data based on its type 
            if current_type == 'Fluid_Temperature': 
                current = data.Fluid_Temperature.data 
            elif current_type == 'Current_Jz': 
                current = data.Current_Jz.data 
            else:  # Fluid Density 
                current = data.Fluid_Rho.data 


            vmin, vmax = vmin_vmax[current_type] 
            if vmin <0: 
                vmin = 0 
                
            if current_type == 'Current_Jz':
                vmax = int(vmax)
                norm = mcolors.PowerNorm(gamma=0.4, vmin=vmin, vmax=vmax) #manually change if required
                title = current_labels[1] 
            elif current_type == 'Fluid_Rho':
                title = current_labels[2] 
                norm = mcolors.PowerNorm(gamma=1, vmin=vmin, vmax=vmax) 
            else:
                title = current_labels[0] 
                norm = mcolors.PowerNorm(gamma=1, vmin=vmin, vmax=vmax) 

            

            # Plotting the 'current' data in the appropriate subplot 
            pcm = axs[i, j].pcolormesh(X, Y, current, norm=norm,cmap=maps[j]) 
            axs[i, j].pcolormesh(X, -Y, current, norm=norm, cmap=maps[j]) 
            
            #axs[i, j].set_title(f"{current_type} at time"+r"$t =$ "+f"{np.round(t,2)}", size=20) 
            #axs[i, j].set_title(fr"$t =$ "+f"{np.round(t,2)}", size=20) 
            
            # Adjusting y-axis limits and labels if necessary 
            axs[i, j].set_ylim([-0.1, 0.1]) 
            if i ==0:
                axs[i, j].set_title(f"{title}", size=25, pad=10) 
            if i == n-1:  # Set x-labels on the last row 
                axs[i, j].set_xlabel(r'$x$', size=25) 
            if j == 0: # Set y-labels on the first column 
                axs[i, j].set_ylabel(r'$z$', size=30) 
                #axs[i, j].text(0.5, 1.05, fr"$t =$ {np.round(t,2)}", size=20, ha='center', va='bottom', transform=axs[i, j].transAxes)
                axs[i, j].text(-0.3, 0.5, fr"$t =$ {np.round(t,2)}", size=22, ha='center', va='center', rotation=90, transform=axs[i, j].transAxes)

            axs[i,j].yaxis.set_major_locator(MaxNLocator(nbins=2))
            
            axs[i,j].set_aspect('auto')
            # Colour bars with labels
            
            divider = make_axes_locatable(axs[i,j])
            cax = divider.append_axes('right', size='5%', pad=0.04)
            cbar = fig.colorbar(pcm, cax=cax, orientation='vertical')
            #cbar.ax.yaxis.set_major_locator(MaxNLocator(nbins=2))
            
            
            
            # Manually set the colorbar ticks to min and max values
            vmin, vmax = vmin_vmax[current_type]
            if current_type == 'Current_Jz':
                cbar.set_label(colorbar_labels[j], size=25, labelpad=-15)  # Setting the colorbar label
                vmax=int(vmax)
                vmin=int(vmin)
                cbar.set_ticks([vmin, vmax])
                cbar.set_ticklabels([f'{vmin}', f'{vmax}'])
            else: 
                cbar.set_label(colorbar_labels[j], size=25, labelpad=-10)  # Setting the colorbar label
                cbar.set_ticks([vmin, vmax])
                cbar.set_ticklabels([f'{vmin:.1f}', f'{vmax:.1f}'])
    
        # Limit the maximum number of colorbar ticks to 2

    # Put colour bar to the right of the plots 
    fig.subplots_adjust(right=0.8, wspace=0.27) 

    #cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7]) 

    #fig.colorbar(pcm, cax=cbar_ax, label='Value') 
    plt.tight_layout() 

    plt.show() 

    return fig 
```
